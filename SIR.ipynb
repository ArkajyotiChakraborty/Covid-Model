$$ s(t+\Delta t) = s(t) - b\Delta t s(t)i(t) $$
$$ r(t+\Delta t) = r(t) + \Delta t k i(t) $$
$$ i(t+\Delta t) = i(t) + \Delta t (bs(t)-k)i(t) $$

```python
import matplotlib.pyplot as plt # Plotting library imported
%matplotlib inline 

```

```python
s = 0.999144
i = 0.00085
r = 0.000006
t = 0
t_final = 125
delta_t = 1.5
N_slices = int(t_final/delta_t)
s_list = [s]
i_list = [i]
r_list = [r]
t_list = [t]
total_fraction = [s+i+r]
```

```python
b = 0.19
k = 0.07


for step in range(0,N_slices):
    s_temp = s - b*delta_t*s*i
    r_temp = r + k*delta_t*i
    i_temp = i + delta_t*(b*s - k)*i
    
    s = s_temp
    r = r_temp
    i = i_temp
    
    t += delta_t
    
    s_list.append(s)
    r_list.append(r)
    i_list.append(i)
    t_list.append(t)
    total_fraction.append(s+r+i)
```

```python
plt.xlabel('t') # Label of 'x' axis
plt.style.use('seaborn-whitegrid')
plt.title('Pandemic progress') # Title of plot
plt.plot(t_list, i_list, linestyle = '-', color='red', label='infected')
plt.plot(t_list, s_list, linestyle = '-', color='orange', label='susceptible')
plt.plot(t_list, r_list, linestyle = '-', color='green', label='recovered')
plt.legend();
```
